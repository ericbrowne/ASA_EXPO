### <u>Merge ACS and SVI: </u>

```python
# read in ACS main:
acsMain = pd.read_csv('estiamtes19_v2.csv',engine='python')
acsMain.head()
```

```python
acsMain.shape
```

```python
## Sample Every 10th row:
acsMain = acsMain.iloc[0::10, :]
acsMain = acsMain.reset_index(drop=True)
acsMain.head()
```

```python
acsMain.shape
```

```python
# Take out the '05000US' out of the GEOID to properly join on SVI
acsMain = acsMain.replace(to_replace='05000US',value='',regex=True)
acsMain.head()
```

```python
acsMain['GeoID'] # worked
```

```python
# Import SVI:
svi = pd.read_csv('SVI2018_US_COUNTY.csv')
print(svi.head())
print(svi.shape)
```

```python
svi.head()
```

```python
svi = svi.drop(['STATE','ST_ABBR','ST'],axis=1)
```

```python
#use pandas str.rjust()
#df['DATA'] = df['DATA'].astype(str)
#df['DATA'] = df['DATA'].str.rjust(6,'0')
svi['FIPS'] = svi['FIPS'].astype(str)
svi['FIPS'] = svi['FIPS'].str.rjust(5,'0')
#svi['FIPS'] = svi['FIPS'].str.zfill(5)
svi['FIPS'] = svi['FIPS'].astype(int)
svi.head()
```

```python
svi['FIPS'] = svi['FIPS'].astype(int)
```

```python
# Rename acsMAin column 'GEOID' to 'FIPS'
acsMain = acsMain.rename(columns={'GeoID':'FIPS'})
acsMain.head()
```

```python
acsMain['FIPS'] = acsMain['FIPS'].astype(int)
```

```python
acsMain.dtypes
```

```python
svi.dtypes
```

```python
print(f'ACS Main Size: {acsMain.shape}')
print(f'SVI Size: {svi.shape}')
# Because the ACS is larger, we will be joining on SVI 
```

```python
acsMain.isnull().sum().sum() # no NA's in any of the col's
```

```python
# check for nulls on SVI as well:
svi.isnull().sum()
```

```python
print(len(acsMain['FIPS'].unique()))
print(len(svi['FIPS'].unique()))
```

```python
acsMain.FIPS
```

```python
svi.FIPS
```

```python
## Try keeping the FIPS as strings:
acsMain['FIPS'] = acsMain['FIPS'].astype(int)
svi['FIPS'] = svi['FIPS'].astype(int)

## check to make sure theyre both strings
print(acsMain.FIPS.dtypes)
print(svi.FIPS.dtypes)
```

```python
# Now try merging on FIPS
acs_svi_merged = acsMain.join(svi, on='FIPS', how='left',lsuffix='_left', rsuffix='_right')
#acs_svi_merged = acsMain.join(svi, on='FIPS', how='left')
acs_svi_merged.head()
```

```python
acs_svi_merged.shape
```

```python
print(len(acsMain.columns))
print(len(svi.columns))
```

```python
# Change column back to "FIPS"
acs_svi_merged.drop('FIPS_right',axis=1,inplace=True)
acs_svi_merged.shape
```

```python
acs_svi_merged = acs_svi_merged.rename(columns={'FIPS_left':'FIPS'})
acs_svi_merged.FIPS
```

```python
# now write to a CSV file
acs_svi_merged.to_csv('acs_svi_merged2.csv')
```
