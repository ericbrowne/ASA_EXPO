```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in ACS main:
acsMain = pd.read_csv('19estimates_counties.csv',engine='python')
acsMain.head()
```

```python
# Take out the '05000US' out of the GEOID to properly join on SVI
acsMain = acsMain.replace(to_replace='05000US',value='',regex=True)
acsMain.head()
```

```python
# Import SVI:
svi = pd.read_csv('SVI2018_US_COUNTY.csv')
print(svi.head())
print(svi.shape)
```

```python
#use pandas str.rjust()
#df['DATA'] = df['DATA'].astype(str)
#df['DATA'] = df['DATA'].str.rjust(6,'0')
svi['FIPS'] = svi['FIPS'].astype(str)
svi['FIPS'] = svi['FIPS'].str.rjust(5,'0')
svi['FIPS'] = svi['FIPS'].astype(int)
```
